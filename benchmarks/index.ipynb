```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head(30)
```

```python
# Matrix Size,NVIDIA GeForce RTX 4060 Laptop GPU, AMD Ryzen 7 7840HS
# Plot the size vs time for both the GPU and CPU
plt.plot(data['Matrix Size'], data['NVIDIA GeForce RTX 4060 Laptop GPU'], label='NVIDIA GeForce RTX 4060 Laptop GPU')
plt.plot(data['Matrix Size'], data['AMD Ryzen 7 7840HS'], label='AMD Ryzen 7 7840HS CPU')
plt.xlabel('Matrix Size')
plt.ylabel('Time (s)')
plt.title('Matrix Size vs Time')
plt.legend()
plt.show()
```

```python
# Get the average acceleration of the GPU
gpu = data['NVIDIA GeForce RTX 4060 Laptop GPU']
cpu = data['AMD Ryzen 7 7840HS']
acceleration = cpu / gpu
average_acceleration = acceleration.mean()
print(f'Average acceleration: {average_acceleration}')
```
